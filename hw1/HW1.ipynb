```python
# Display all pitchers in 2015 sorted by their ERAs. This allows you to identify the best and worst pitchers.
pitching.sort_values(by = 'ERA')
```

```python
# Display all pitchers in 2015 under each team sorted by their strike outs (SO).
pitching.groupby('teamID').apply(lambda x: x.sort_values('SO'))
```

```python
# Calculate the average ERA for each team, list the teams by the average ERA in a descending order.
a = pd.DataFrame({'ERA' : pitching.groupby(["teamID"])["ERA"].mean()})
a.sort_values(by=['ERA'], ascending=False)
```

```python
# Calculate the total number of home runs allowed (HRs) for each team, and list the team by the count in an ascending order.
b = pd.DataFrame({'HRs' : pitching.groupby(["teamID"])["HR"].sum()})
b.sort_values(by=['HRs'])
```

```python
# Calculate each team’s average ERA from 2011 to 2015, and list the result as a table, where each column is a year, each row is a team.
t1 = pitching.loc[pitching['yearID'] == 2011]
t2 = pitching.loc[pitching['yearID'] == 2012]
t3 = pitching.loc[pitching['yearID'] == 2013]
t4 = pitching.loc[pitching['yearID'] == 2014]
t5 = pitching.loc[pitching['yearID'] == 2015]
pd.DataFrame({'2011' : t1.groupby(["teamID"])["ERA"].mean(), '2012' : t2.groupby(["teamID"])["ERA"].mean(), '2013' : t3.groupby(["teamID"])["ERA"].mean(), '2014' : t4.groupby(["teamID"])["ERA"].mean(), '2015' : t5.groupby(["teamID"])["ERA"].mean()})
```

```python
# Calculate the average ERAs for American League (AL) and National League (NL) from 2011 to 2015, and list the result as a table, where each column is a league and each row is a year.
c = pitching.loc[(pitching['yearID'] >= 2011) & (pitching['yearID'] <= 2015)]
AL = c.loc[c['lgID'] == 'AL']
NL = c.loc[c['lgID'] == 'NL']
pd.DataFrame({'AL' : AL.groupby(["yearID"])["ERA"].mean(), 'NL' : NL.groupby(["yearID"])["ERA"].mean()})
```

```python
# List the pitchers who had the most wins in each of the years from 2011 to 2015.
t1 = pitching.loc[pitching['yearID'] == 2011]
t2 = pitching.loc[pitching['yearID'] == 2012]
t3 = pitching.loc[pitching['yearID'] == 2013]
t4 = pitching.loc[pitching['yearID'] == 2014]
t5 = pitching.loc[pitching['yearID'] == 2015]
t1 = t1[['yearID', 'nameFirst', 'nameLast', 'W']][t1.W == t1.W.max()]
t2 = t2[['yearID', 'nameFirst', 'nameLast', 'W']][t2.W == t2.W.max()]
t3 = t3[['yearID', 'nameFirst', 'nameLast', 'W']][t3.W == t3.W.max()]
t4 = t4[['yearID', 'nameFirst', 'nameLast', 'W']][t4.W == t4.W.max()]
t5 = t5[['yearID', 'nameFirst', 'nameLast', 'W']][t5.W == t5.W.max()]
pd.concat([t1, t2, t3, t4, t5], axis = 0).set_index('yearID') 
```

```python
players = pd.read_csv('players.csv')  
teams = pd.read_csv('teams.csv') 
```

```python
# Compute the total number of wins for each of the teams in MLB over 2011-2015, sort them in a descending order.
t1 = teams.loc[(teams['yearID'] >= 2011) & (teams['yearID'] <= 2015)]
pd.DataFrame({'W' : t1.groupby(["teamID"])["W"].sum()}).sort_values(by=['W'], ascending=False)
```

```python
# Compute the average payroll per year for all teams over 2011-2015, sort them in a descending order.
t1 = players.loc[players['yearID'] == 2011]
t2 = players.loc[players['yearID'] == 2012]
t3 = players.loc[players['yearID'] == 2013]
t4 = players.loc[players['yearID'] == 2014]
t5 = players.loc[players['yearID'] == 2015]
t1 = pd.DataFrame({'payroll' : t1.groupby(["teamID"])["salary"].mean()})
t1 = t1.rename(index = lambda x: x + ' 2011')
t2 = pd.DataFrame({'payroll' : t2.groupby(["teamID"])["salary"].mean()})
t2 = t2.rename(index = lambda x: x + ' 2012')
t3 = pd.DataFrame({'payroll' : t3.groupby(["teamID"])["salary"].mean()})
t3 = t3.rename(index = lambda x: x + ' 2013')
t4 = pd.DataFrame({'payroll' : t4.groupby(["teamID"])["salary"].mean()})
t4 = t4.rename(index = lambda x: x + ' 2014')
t5 = pd.DataFrame({'payroll' : t5.groupby(["teamID"])["salary"].mean()})
t5 = t5.rename(index = lambda x: x + ' 2015')
pd.concat([t1, t2, t3, t4, t5], axis = 0).sort_values(by=['payroll'], ascending=False)
```

```python
# Create a visualization of your choice which will all allow you to show whether a team’s winning record is related to its payroll. 
t1 = players.loc[players['yearID'] == 2011]
t2 = players.loc[players['yearID'] == 2012]
t3 = players.loc[players['yearID'] == 2013]
t4 = players.loc[players['yearID'] == 2014]
t5 = players.loc[players['yearID'] == 2015]
T1 = teams.loc[teams['yearID'] == 2011, ['W','teamID']]
T2 = teams.loc[teams['yearID'] == 2012, ['W','teamID']]
T3 = teams.loc[teams['yearID'] == 2013, ['W','teamID']]
T4 = teams.loc[teams['yearID'] == 2014, ['W','teamID']]
T5 = teams.loc[teams['yearID'] == 2015, ['W','teamID']]
t1 = pd.DataFrame({'payroll' : t1.groupby(["teamID"])["salary"].mean()})
t1 = pd.merge(t1, T1, on = 'teamID').set_index('teamID') 
t1 = t1.rename(index = lambda x: x + ' 2011')
t2 = pd.DataFrame({'payroll' : t2.groupby(["teamID"])["salary"].mean()})
t2 = pd.merge(t2, T2, on = 'teamID').set_index('teamID') 
t2 = t2.rename(index = lambda x: x + ' 2012')
t3 = pd.DataFrame({'payroll' : t3.groupby(["teamID"])["salary"].mean()})
t3 = pd.merge(t3, T3, on = 'teamID').set_index('teamID') 
t3 = t3.rename(index = lambda x: x + ' 2013')
t4 = pd.DataFrame({'payroll' : t4.groupby(["teamID"])["salary"].mean()})
t4 = pd.merge(t4, T4, on = 'teamID').set_index('teamID') 
t4 = t4.rename(index = lambda x: x + ' 2014')
t5 = pd.DataFrame({'payroll' : t5.groupby(["teamID"])["salary"].mean()})
t5 = pd.merge(t5, T5, on = 'teamID').set_index('teamID') 
t5 = t5.rename(index = lambda x: x + ' 2015')
df = pd.concat([t1, t2, t3, t4, t5], axis = 0)
df = df.reset_index()
df[['Team', 'Year']] = df.teamID.str.split(expand=True) 
Team = list(df['Team'][:30])
import matplotlib.pyplot as plt    
for i in range(30):
    x = df[df['Team'] == Team[i]]['W']
    y = df[df['Team'] == Team[i]]['payroll']
    plt.scatter(x, y, label = Team[i]) 
plt.xlabel("Win")
plt.ylabel("payroll")
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.show()       
```

Q : So is a team’s winning record related to its payroll?
A : 根據圖表，勝率特別高的幾支球隊反而整體 payroll 是不到 8M 的，因此勝率與 payroll 並無直接相關。

```python
# Compute the Batting Averages for each of the MLB teams over 2011-2015, sort them in a descendin- gorder. The Batting Average is defined as Hits/At Bats. The average is calculated from all players in each team.
t = players.loc[(players['yearID'] >= 2011) & (players['yearID'] <= 2015)]
T = pd.DataFrame(index = ['avg'], columns = Team)
k = 0
for i in Team:
    T.iloc[0, k] = t.loc[(t['teamID'] == i)]['H'].sum()/t.loc[(t['teamID'] == i) ]['AB'].sum()
    k+=1
T = T.T # transpose
T.sort_values(by=['avg'], ascending = 'False')
```

```python
# Create a visualization of your choice which will allow you to decide whether a team’s batting average is related to its win-loss record. 
tem = teams.loc[(teams['yearID'] >= 2011) & (teams['yearID'] <= 2015)]
tt = list()
for i in Team:
    tt.append(tem.loc[tem['teamID'] == i]['W'].sum())
T['Win'] = tt
for i in range(30):
    x = T.iloc[i,0]
    y = T.iloc[i,1]
    plt.scatter(x, y, label = T.index[i]) 
plt.xlabel("batting average")
plt.ylabel("Win")
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.show()     
```

Q : So is a team’s winning record related to its batting average?
A : 根據圖表，大部分的球隊勝場數都在400多勝，而打擊率最高的底特律老虎隊勝場數也不是最高的，因此勝場數與打擊率也無關。

```python
# Display the average ERA (Earned Run Average) for each of the MLB teams in 2011-2015, sort them in a descending order. A lower ERA indicates a better pitching performance.
t1 = pitching.loc[(pitching['yearID'] >= 2011) & (pitching['yearID'] <= 2015)]
tt = list()
for i in Team:
    tt.append(t1.loc[t1['teamID'] == i]['ERA'].mean())
T['ERA'] = tt
T.sort_values(by=['ERA'])
```

```python
# Create a visualization of your choice which will allow you to decide if a team’s win-loss record is related to its pitching performance.
for i in range(30):
    x = T.iloc[i,2]
    y = T.iloc[i,1]
    plt.scatter(x, y, label = T.index[i]) 
plt.xlabel("ERA")
plt.ylabel("Win")
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.show()    
```

Q : So is a team’s winning record related to its pitching performance?
